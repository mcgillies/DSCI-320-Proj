```python
import pandas as pd
import altair as alt
import matplotlib as plt
```

```python
base = pd.read_csv("data/NBA2000_2022_V1.csv")
base.head()
```

```python
base_filtered = base.drop(columns = ['Unnamed: 0', '...10', '...17', '...24', '...27', '...30', '...33', 'Abbreviation', 
                                    'Rk', 'Pos', 'Age', 'Tm', 'G', "MP"])
base_filtered[(base_filtered['name'] == 'Jamal Crawford')]
```

```python
numeric = base_filtered.drop(columns = ["slug", "positions", 'team'])
cat = base_filtered[["slug", "positions", 'team', 'name']]
```

```python
numeric_grouped = numeric.groupby(by = 'name').mean()
numeric_grouped.head()
#base_grouped[base_grouped['name'] == 'Kevin Durant']
```

```python
cat_grouped = cat.groupby("name").agg(pd.Series.mode)
cat_grouped.head()
```

```python
grouped_total = cat_grouped.merge(numeric_grouped, on = 'name')
grouped_total.head()
```

```python
grouped_total.to_csv("data/Averages.csv")
```

```python
char = pd.read_csv("data/all_seasons.csv")
char = char.drop(columns = 'Unnamed: 0')
char['season'] = char['season'].str[:4].astype(int) + 1
char = char.rename(columns = {'player_name':'name', 'season':'Year'})
char[(char['name'] == 'Jamal Crawford')]
```

```python
char_years = char[(char['Year'] > 2000)]
char_years['Year'].value_counts()
```

```python
merged_player = char_years.merge(base_filtered, on = ['name', 'age'])
merged_player = merged_player.drop(columns = ['gp', 'team_abbreviation'])
merged_player.head()
merged_player[(merged_player['name'] == 'Jamal Crawford')]
```

```python
merged_player.columns
```

```python
merged_player.to_csv("data/PlayerCharacteristics.csv")
```
